```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
from imblearn.over_sampling import SMOTE
```

```python
df=pd.read_csv('wine-class.csv')
```

```python
X=df.drop(columns=['class'])
y=df['class']
```

```python
print("Input samples:\n",X.head())
print("Output Samples:\n",y.head())
```

```python
print("Class Distribution:\n",df['class'].value_counts())
```

```python
# Example: clean a string line
line = "X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42, stratify=y)"
line = line.replace('\xa0', ' ')
print(line)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42, stratify=y)
```

```python
scaler=StandardScaler()
X_train,X_test=scaler.fit_transform(X_train),scaler.transform(X_test)
```

```python
print(pd.DataFrame(X_train).head())
```

```python
X_train_res,y_train_res=SMOTE(random_state=42).fit_resample(X_train,y_train)
```

```python
print("Class Distribution:\n",pd.Series(y_train_res).value_counts().sort_index())
```

```python
lr=LogisticRegression(max_iter=5000,multi_class='multinomial')
lr.fit(X_train_res,y_train_res)
```

```python
print("Logistic Regression \n",classification_report(y_test,lr.predict(X_test)))
```

```python
rf=RandomForestClassifier(class_weight='balanced',random_state=42)
rf.fit(X_train_res,y_train_res)
```

```python
print("Random Forest \n",classification_report(y_test,lr.predict(X_test)))
```
